```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as py
%matplotlib notebook
%matplotlib inline

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_data = pd.merge(mouse_metadata,study_results,on=["Mouse ID"])

# Display the data table for preview
combined_data
```

```python
# Checking the number of mice.
print(combined_data["Mouse ID"].nunique())
print(combined_data["Mouse ID"].count())
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data=combined_data.drop_duplicates(subset=["Mouse ID","Timepoint"])
clean_data
```

```python
# Checking the number of mice in the clean DataFrame
print(clean_data["Mouse ID"].nunique())
print(clean_data["Mouse ID"].count())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method is the most straighforward, creating multiple series and putting them all together at the end.
drug_group = clean_data.groupby("Drug Regimen")
long_mean = round(drug_group["Tumor Volume (mm3)"].mean(),2)
long_median = round(drug_group["Tumor Volume (mm3)"].median(),2)
long_var = round(drug_group["Tumor Volume (mm3)"].var(),2)
long_std = round(drug_group["Tumor Volume (mm3)"].std(),2)
long_sem = round(drug_group["Tumor Volume (mm3)"].sem(),2)
long_summary_table = pd.DataFrame({
    "Mean":long_mean, "Median":long_median,"Variance":long_var,
    "Standard Deviation":long_std,"SEM":long_sem
})
long_summary_table
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method produces everything in a single groupby function
summary_stats=round(drug_group["Tumor Volume (mm3)"].agg(["mean","median","var","std","sem"]),2)
summary_stats=summary_stats.rename(columns={"mean":"Mean","median":"Median","var":"Variance",
                                            "std": "Standard Deviation","sem":"SEM"})
summary_stats
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of mice for each treatment throughout the course of the study using pandas. 
mouse_count=drug_group["Mouse ID"].nunique()
mouse_count.plot(kind="bar",title="Number of Mice per Drug Type", color="green",alpha=.6, 
                 ylim=(0,mouse_count.max()+2),width=.6)
plt.xticks(rotation=45)
plt.ylabel("Number of Mice")
plt.show()
```

```python
# Generate a bar plot showing the total number of mice for each treatment throughout the course of the study using pyplot.
plt.bar(mouse_count.index, mouse_count, color="green",alpha=.6, width=.6)
plt.xticks(rotation=45)
plt.ylim(0,mouse_count.max()+2)
plt.title("Number of Mice per Drug Type")
plt.ylabel("Number of Mice")
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pandas
gender_count=clean_data["Sex"].value_counts()
gender_count.plot(kind="pie", title="Percent Gender",autopct="%1.2f%%",colors=["yellow","magenta"],legend=True,labels=None)
plt.ylabel("")
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot
plt.pie(gender_count,colors=["yellow","magenta"],autopct="%1.2f%%",labels=["Male","Female"])
plt.title("Percent Gender")
plt.legend()
plt.show()
```
