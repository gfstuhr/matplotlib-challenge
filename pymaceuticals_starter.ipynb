```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as sts
import numpy as py
%matplotlib notebook
%matplotlib inline

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_data = pd.merge(mouse_metadata,study_results,on=["Mouse ID"])

# Display the data table for preview
combined_data
```

```python
# Checking the number of mice.
print(combined_data["Mouse ID"].nunique())
print(combined_data["Mouse ID"].count())
```

```python
dupes = combined_data.loc[combined_data.duplicated(subset=["Mouse ID","Timepoint"]),
                          "Mouse ID"].unique()
print(dupes)
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
combined_data.loc[combined_data.loc[:,"Mouse ID"]== "g989",:]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data=combined_data.drop_duplicates(subset=["Mouse ID","Timepoint"])
clean_data
```

```python
# Checking the number of mice in the clean DataFrame
print(clean_data["Mouse ID"].nunique())
print(clean_data["Mouse ID"].count())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method is the most straighforward, creating multiple series and putting them all together at the end.
drug_group = clean_data.groupby("Drug Regimen")
mean = round(drug_group["Tumor Volume (mm3)"].mean(),2)
median = round(drug_group["Tumor Volume (mm3)"].median(),2)
var = round(drug_group["Tumor Volume (mm3)"].var(),2)
std = round(drug_group["Tumor Volume (mm3)"].std(),2)
sem = round(drug_group["Tumor Volume (mm3)"].sem(),2)
summary_table = pd.DataFrame({
    "Mean":mean, "Median":median,"Variance":var,
    "Standard Deviation":std,"SEM":sem
})
summary_table
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method produces everything in a single groupby function
summary_stats=round(drug_group["Tumor Volume (mm3)"].agg(["mean","median","var","std","sem"]),2)
summary_stats=summary_stats.rename(columns={"mean":"Mean","median":"Median","var":"Variance",
                                            "std": "Standard Deviation","sem":"SEM"})
summary_stats
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
max_time=clean_data.groupby("Mouse ID").max()["Timepoint"]

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
max_time_study = pd.merge(max_time,clean_data,on=["Mouse ID","Timepoint"])
max_time_study
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drug_list=["Capomulin","Ramicane","Infubinol","Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol=[]
four_quantiles=[.25,.5,.75]

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for each_drug in drug_list:
    drug_df=max_time_study.loc[:,"Drug Regimen"] == each_drug
    drug_tumor_vol=max_time_study.loc[drug_df,"Tumor Volume (mm3)"]
    tumor_vol.append(drug_tumor_vol)
    quartiles=drug_tumor_vol.quantile(four_quantiles)
    lowerq=quartiles.loc[.25]
    upperq=quartiles.loc[.75]
    iqr=upperq-lowerq
    lower_bound=lowerq-(1.5*iqr)
    upper_bound=upperq+(1.5*iqr)
   
    # Locate the rows which contain mice on each drug and get the tumor volumes   
    # add subset 
    # Determine outliers using upper and lower bounds
    outlier_df=(drug_tumor_vol>upper_bound)|(drug_tumor_vol<lower_bound)
    outlier_count=len(drug_tumor_vol.loc[outlier_df])
    print(f"There are {outlier_count} outliers for {each_drug}")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
plt.boxplot(tumor_vol, labels=drug_list)
plt.title("Box Plot for each Drug")
plt.xlabel("Drug Regimen")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
mouse_df=clean_data["Drug Regimen"]=="Capomulin"
capomulin=clean_data[mouse_df]
mouse_df=capomulin["Mouse ID"]=="v923"
mouse=capomulin[mouse_df]
mouse.plot("Timepoint", "Tumor Volume (mm3)", title="Tumor Volume Over Time", legend=False,
          linewidth=3, color="red",alpha=.5)
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
capomulin_avg=capomulin[["Mouse ID","Tumor Volume (mm3)","Weight (g)"]].groupby("Mouse ID").mean()
capomulin_avg.plot.scatter("Weight (g)","Tumor Volume (mm3)", title="Mouse Weight vs Tumor Volume",
                          color="red",alpha=.5, marker="*")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
corr=round(sts.pearsonr(capomulin_avg["Weight (g)"],capomulin_avg["Tumor Volume (mm3)"])[0],2)
print(f"The correlation is {corr}.")
```

```python
#Linear regression model
slope, intercept, rvalue, pvalue, stderr=sts.linregress(capomulin_avg["Weight (g)"],
                                                        capomulin_avg["Tumor Volume (mm3)"])
y_line=slope*capomulin_avg["Weight (g)"]+intercept
capomulin_avg.plot.scatter("Weight (g)","Tumor Volume (mm3)", title="Mouse Weight vs Tumor Volume",
                          color="red",alpha=.5, marker="*")
plt.plot(capomulin_avg["Weight (g)"],y_line, color="green", alpha=.5)
plt.show()
```
